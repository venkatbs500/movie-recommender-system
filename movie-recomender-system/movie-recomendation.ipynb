```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
movies = movies.merge(credits,on='title') # we merged 2 datasets soo that is will be easier
```

```python
movies.head()
```

```python
#since we hav a lot of data which is not needed we can just remove those set of column and make a new data set accordingly
#we are gonna use: id, title, overview, genere, keywords, cast, crew
```

```python
movies.info()
```

```python
movies = movies[['id', 'original_title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
#we need only the name in the dict soo we need a function
```

```python
import ast #imported this too convert to list
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
# for cast we need only the names of the top 3 actors soo we use this function
```

```python
import ast #imported this too convert to list
def convert3(obj):
    L=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
 movies.head()
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
#we will convert the overview to the list which is in string to easily use it with others

movies['overview']=movies['overview'].apply(lambda x: x.split())
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
#the next 4 lines of code are to remove the spaces between the words
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['id', 'original_title', 'tags', 'overview']]
```

```python
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
 new_df.head()
```

```python
 new_df['tags']=new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

```python
 import nltk
```

```python
from nltk.stem.porter import PorterStemmer 
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
#vectorization
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray() 
```

```python
vectors
```

```python
vectors[0]
```

```python
 from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors )
```

```python
similarity
```

```python
sorted(list(enumerate(similarity[0])),reverse = True,key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
        movie_index = new_df[new_df['original_title'] == movie].index[0]
        distances = similarity[movie_index]
        movies_list = sorted(list(enumerate(distances)),reverse = True,key = lambda x:x[1])[1:6]
        
        for i in movies_list:
             print(new_df.iloc[i[0]].original_title)
        
```

```python
recommend('Avatar')
```
